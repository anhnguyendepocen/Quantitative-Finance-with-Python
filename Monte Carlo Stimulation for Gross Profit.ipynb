# Introduction

We are interested in different possible realizations/scenarios of a "future event".
We can use past data to create a simulation-a new larger set of fictional but sensible data.
This stimulation takes into account the mean and variance of historical data and creates a larger dataset according to historical data distribution.
It is a good proxy for different outcome.
It is used in corporate finance, investment evaluation, asset and risk management, estimating insurance liabilities, pricing of options and other derivatives.

The high uncertainty in finance makes MCS a valuable tool that improves decision-making process when several random variables are in play.

Finance managers forcast revenues, cogs and opex.

Example of monte carlo stimulation

Current Year Revenue= Last year's revenue(1+growth rate)

We will run 1000 stimulations for different growth rates and calculate average, max and min expected current year revenue.


Formulae

Gross Profit = Revenue - COGS

Operating Profit = Gross Profit - Opex





# Predicting future Gross Profit

We will need exp revenue and exp cogs.

We will run 1000 stimulations for exp revenue.

Suppose exp revenue is $170  with stdv $20

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rev_m=170
rev_stdev=20
iterations=1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
```

```python
rev
```

```python
rev.shape
```

```python
plt.figure(figsize=(15,6)),
plt.plot(rev)
plt.show()
```

```python
#suppose COGS is approx 60% of revenue with normal distribution mean=0.6 and stdev=0.1
#COGS Should be negative because it is the money spent

cogs = -(rev * np.random.normal(0.6,0.1))#no iternations included because rev object already has that.
cogs.shape
```

```python
plt.figure(figsize=(15,6));
plt.plot(cogs)
plt.show()
```

```python
cogs.mean()
```

```python
cogs.std()
```

```python
Gross_Profit= rev + cogs#positive sign because before that we used negative sign. GP=Rev-Cost
Gross_Profit
```

```python
plt.figure(figsize=(15,10));
plt.plot(Gross_Profit)
plt.show()
```

```python
max(Gross_Profit)
```

```python
min(Gross_Profit)
```

```python
Gross_Profit.mean()
```

```python
Gross_Profit.std()
```
